#b .To create boxplots and identify outliers and Inter Quarter Range

#numerical columns
1.volume , 2.avg price , 3.total sales ,4.discount rate ,5.discount amount ,6.net values value

```python
import pandas as pd
data = pd.read_csv('/content/sales_data_with_discounts.csv')
data.head()
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

```

```python
plt.boxplot(data['Volume'])
```

```python
Q1 = data['Volume'].quantile(0.25)
print(Q1)
Q3 = data['Volume'].quantile(0.75)
print(Q3)
IQR = Q3 -Q1
print(IQR)
print('Upper limit is UL = ', Q3 + (1.5 * IQR))
print('Lower Limit is LL = ', Q1 - (1.5 * IQR))
```

```python
outliers = data[(data['Volume'] > 10.5) | (data['Volume'] < -1.5)]
outliers
```

```python
outliers['Volume'].count()
```

for analysing of volume column,we get IQR = 3 and outliers is 44

--------------------------------*-----------------------------

```python
plt.boxplot(data['Avg Price'])
```
